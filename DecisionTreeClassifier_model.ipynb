```python
y=weather['high_humidity_3pm']
X=weather.drop(['high_humidity_3pm'],axis=1)
```

```python
dec_tree=DecisionTreeClassifier()
```

```python
dec_tree.fit(X_train,y_train)
```

```python
y_predict=dec_tree.predict(X_test)
y
```

```python
accuracy_score(y_test,y_predict)
```
